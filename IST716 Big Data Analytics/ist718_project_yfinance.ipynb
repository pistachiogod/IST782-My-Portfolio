```python

import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta


# Convert 'Stock' column to string to ensure all entries are strings
new_df['Cleaned_Stocks'] = new_df['Cleaned_Stocks'].astype(str)

def get_stock_price_on_date(symbol, date):
    # Ensure symbol is a string; this also converts NaN and None to 'nan'
    symbol = str(symbol)
    if symbol.lower() == 'nan':  # Check if the symbol is effectively NaN
        return None
    date_formatted = date.strftime('%Y-%m-%d')
    end_date = date + timedelta(days=1)
    end_date_formatted = end_date.strftime('%Y-%m-%d')
    try:
        stock = yf.Ticker(symbol)
        hist = stock.history(start=date_formatted, end=end_date_formatted)
        if not hist.empty:
            return hist['Close'].iloc[0]
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
    return None

# Ensure 'Parsed_Date' and 'Published_Clean' are datetime objects; convert them if necessary
# df['Parsed_Date'] = pd.to_datetime(df['Parsed_Date'])
# df['Published_Clean'] = pd.to_datetime(df['Published_Clean'])

# Use 'apply' with a lambda function to iterate over rows for both 'Traded' and 'Published' dates
new_df['Price_On_Traded'] = new_df.apply(lambda row: get_stock_price_on_date(row['Cleaned_Stocks'], row['Parsed_Date']), axis=1)
new_df['Price_On_Published'] = new_df.apply(lambda row: get_stock_price_on_date(row['Cleaned_Stocks'], row['Published_Clean']), axis=1)

```
